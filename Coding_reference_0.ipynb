```python
import numpy as np

import scipy.special as spsp

import matplotlib.pyplot as plt

%matplotlib inline

import sympy

import scipy.stats as spst
import simpy 
```

```python
def arrivals(env,stop1,data,desk):
  i=0
  lmbda=100
  people_age=0
  while True:
        inter_arrival=-1/lmbda*np.log(np.random.rand())
        yield env.timeout(inter_arrival)
        i+=1
        
        if np.random.rand()<0.7:
            people_age=1 #young 
        else:
            people_age=0 #old 
        env.process(service(i,env,stop1,data,people_age,desk ))
          
        
    
def service(i,env,stop1,data,people_age,desk ):
    #requesting the server
    if people_age==1:
        rqt=desk.request(priority=-1, preempt=False)
      
    else:
        rqt=desk.request(priority=1,preempt=False)
      
    #occupy/process the server request
    yield rqt
    
    yield env.timeout(0.003472)
    desk.release(rqt)
  
    data['n']-=1
    if env.now>=24 or data['n']==0:
      stop1.succeed()

def system1():
    
    env=simpy.Environment()
    desk=simpy.PreemptiveResource(env,capacity=1000)
    stop1=env.event()
    data={'n':10000}
    env.process(arrivals(env,stop1,data,desk))
    env.run(until=stop1)
    vaccine_left=data['n']
   
    return vaccine_left

```

```python
system1()
```
